### Introduction


Real-world data rarely comes clean. In this project, I'll gather data from various sources , assess its quality and tidiness issues, then clean it. The dataset to be assessed and cleaned is the tweet archive of Twitter user @dog_rates.

### Gathering Data


Data to be gathered will come from 3 sources.

Firstly, the twitter_archive dataframe will be download from classroom page directly.

The second which is image_predictions dataframe will be recovered from udacity cloud. I used request and response to gather the data. Point to note is that it is in form of tab separated values.

The third dataframe was obtained using twitter API. I first downloaded the favorite_count, retweets_count, full_text and lang into a tweet_json.txt file before transferring it into a dataframe.

### Accessing data


I accessed the data visually and programmatically.

Visually, I opened the dataframes one by one and checked for errors that I could find. I also used the sample method to open random rows.

Programmatically, I used various methods such as info, value_counts(), sort_values(), duplicated(), isnull(), notnull(), sum(), count(), list(), dtype, types() and describe().
